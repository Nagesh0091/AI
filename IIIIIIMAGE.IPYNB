```python
thumbnails = Driver.find_elements(By.CLASS_NAME, "MediaCard_overlay__HNJfd")
wait = WebDriverWait(Driver, 10)

for thumbnail in thumbnails:
    try:
        # Scroll element into view
        Driver.execute_script("arguments[0].scrollIntoView(true);", thumbnail)
        time.sleep(1)
        
        # Move to element
        actions.move_to_element(thumbnail).perform()
        time.sleep(1)

        # Wait for download button to be clickable before clicking
        download_button = wait.until(
            EC.element_to_be_clickable((By.CSS_SELECTOR, DOWNLOAD_BUTTON_SELECTOR))
        )
        Driver.execute_script("arguments[0].click();", download_button)
        time.sleep(2)

        # click on empty padding area to close modal
        overlay = wait.until(
            EC.element_to_be_clickable((By.CSS_SELECTOR, "div.Modal_contentPaddingWrapper__zFx4j"))
        )
        Driver.execute_script("arguments[0].click();", overlay)

        time.sleep(1)
    except Exception as e:
        print(f"Error processing thumbnail: {e}")
        continue
```
