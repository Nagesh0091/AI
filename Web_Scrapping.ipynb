```python
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# Wait for all tabs with the target class to be present
tabs = WebDriverWait(driver, 10).until(
    EC.presence_of_all_elements_located((By.CLASS_NAME, "R1QWuf"))
)

# Click the 3rd tab (index 2), which typically is 'Images'
tabs[2].click()
```

```python
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
import time

# Collect all thumbnail elements
thumbnails = driver.find_elements(By.CSS_SELECTOR, "img.rg_i")

img_urls = []

for thumb in thumbnails:
    try:
        # Click the thumbnail (to open high-res preview)
        thumb.click()
        time.sleep(1)  # Wait for the preview to load

        # Locate the large image in the right-side panel
        images = driver.find_elements(By.CSS_SELECTOR, "img.n3VNCb")
        for image in images:
            src = image.get_attribute("src")
            if src and "http" in src:
                img_urls.append(src)
                break  # Only take the first found for each preview
    except Exception as e:
        continue

# Print or save full-res URLs
for url in img_urls:
    print(url)
```
